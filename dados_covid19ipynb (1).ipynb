```python
df.rename(columns={
          'Direction' : 'Direção',
          'Year' : 'Ano',
          'Date' : 'Data',
          'Weekday' : 'Dia da Semana',
          'Country' : 'País',
          'Commodity': 'Mercadoria',
          'Transport_Mode' : 'Modo de Transporte',
          'Measure' : 'Medida',
           'Value' : 'Valor',
          'Cumulative' : 'Acumulado'


}, inplace=True)

display(df.columns)
```

```python
df ['País'] = df['País'].replace({
    'All': 'Todos',
    'China': 'China',
    'East Asia (excluding China)': 'Leste Asiático',
    'United States': 'Estados Unidos',
    'Australia': 'Austrália',
    'United Kingdom': 'Reino Unido',
    'Japan': 'Japão',
    'European Union (27)': 'União Europeia (27)',
    'Total (excluding China)': 'Total (excluindo China)',

})

print(df['País'].unique())
```

```python
df['Mercadoria'] = df['Mercadoria'].replace({
    'All': 'Todas',
    'Milk powder, butter, and cheese': 'Leite em pó, manteiga e queijo',
    'Meat and edible offal': 'Carne e miudezas comestíveis',
    'Logs, wood, and wood articles': 'Toros, madeira e artigos de madeira',
    'Fish, crustaceans, and molluscs': 'Peixes, crustáceos e moluscos',
    'Non-food manufactured goods': 'Bens manufaturados não alimentícios',
    'Electrical machinery and equip': 'Máquinas e equipamentos elétricos',
    'Mechanical machinery and equip': 'Máquinas e equipamentos mecânicos',
    'Fruit': 'Frutas'

})

print(df['Mercadoria'].unique())
```

```python
df['Modo de Transporte'] = df['Modo de Transporte'].replace({
    'All': 'Todos(as)',
    'Air': 'Avião',
    'Sea': 'Navio',
})

print(df['Modo de Transporte'].unique())
```

```python
df['Medida'] = df['Medida'].replace({
    '$' : 'Dólares',
    'Tonnes' : 'Toneladas'
})
print(df['Medida'].unique())
```

```python
df['Medida'] = df['Medida'].replace({
    'All': 'Todos(as)',
    'Value': 'Valor',
    'Weight': 'Peso'
})

print(df['Medida'].unique())
```

```python
df['Dia da Semana'] = df['Dia da Semana'].replace({
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo',
})

print(df['Dia da Semana'].unique())
```

```python
df['Direção'] = df['Direção'].replace({
    'Imports': 'Importações',
    'Exports': 'Exportações',
    'Reimports': 'Reimportações',

  })

print(df['Direção'].unique())
```

```python
df_limpo = df.dropna()
```

```python
df_limpo.isnull().sum()
```

```python
import plotly.express as px
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Calcular média e ordenar
valor_gasto = (
    df_limpo.groupby('País')['Valor']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

# Criar gráfico
fig = px.bar(
    valor_gasto,
    x='País',
    y='Valor',
    title='Valor gasto no total em USD(Dólares)',
    labels={
        'País': 'País',
        'Valor': 'Valor Gasto no Total(Dólares)' # Adjusted label as Medida can be Dólares or Toneladas
    }
)

fig.show()
```

```python
# Count the occurrences of each 'Mercadoria' for each 'País'
mercadorias_counts = (
    df_limpo.groupby(['País', 'Mercadoria'])
    .size()
    .reset_index(name='Count')
)

# Find the top mercadorias for each country (this might result in multiple mercadorias per country if they have the same count)
top_mercadorias_per_country = mercadorias_counts.loc[mercadorias_counts.groupby('País')['Count'].idxmax()]

# Create graph
fig = px.bar(
    top_mercadorias_per_country,
    x='País',
    y='Count',
    color='Mercadoria',  # Use color to distinguish mercadorias
    title='Top Mercadoria por País',
    labels={
        'País': 'País',
        'Count': 'Contagem de Mercadorias',
        'Mercadoria': 'Mercadoria'
        ''
    }
)

fig.show()
```

```python
# Count the occurrences of each 'Mercadoria' across all countries
mercadorias_total_counts = (
    df_limpo.groupby('Mercadoria')
    .size()
    .reset_index(name='Total Count')
)

# Display the total counts
print("Total count of each Mercadoria across all countries:")
display(mercadorias_total_counts)
```
