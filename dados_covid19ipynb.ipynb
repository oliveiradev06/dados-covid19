```python
df.rename(columns={
          'Direction' : 'Direção',
          'Year' : 'Ano',
          'Date' : 'Data',
          'Weekday' : 'Dia da Semana',
          'Country' : 'País',
          'Commodity': 'Mercadoria',
          'Transport_Mode' : 'Modo de Transporte',
          'Measure' : 'Medida',
           'Value' : 'Valor',
          'Cumulative' : 'Acumulado'


}, inplace=True)

display(df.columns)
```

```python
df ['País'] = df['País'].replace({
    'All': 'Todos',
    'China': 'China',
    'East Asia (excluding China)': 'Leste Asiático',
    'United States': 'Estados Unidos',
    'Australia': 'Austrália',
    'United Kingdom': 'Reino Unido',
    'Japan': 'Japão',
    'European Union (27)': 'União Europeia (27)',
    'Total (excluding China)': 'Total (excluindo China)',

})

print(df['País'].unique())
```

```python
df['Mercadoria'] = df['Mercadoria'].replace({
    'All': 'Todas',
    'Milk powder, butter, and cheese': 'Leite em pó, manteiga e queijo',
    'Meat and edible offal': 'Carne e miudezas comestíveis',
    'Logs, wood, and wood articles': 'Toros, madeira e artigos de madeira',
    'Fish, crustaceans, and molluscs': 'Peixes, crustáceos e moluscos',
    'Non-food manufactured goods': 'Bens manufaturados não alimentícios',
    'Electrical machinery and equip': 'Máquinas e equipamentos elétricos',
    'Mechanical machinery and equip': 'Máquinas e equipamentos mecânicos',
    'Fruit': 'Frutas'

})

print(df['Mercadoria'].unique())
```

```python
df['Modo de Transporte'] = df['Modo de Transporte'].replace({
    'All': 'Todos(as)',
    'Air': 'Avião',
    'Sea': 'Navio',
})

print(df['Modo de Transporte'].unique())
```

```python
df['Medida'] = df['Medida'].replace({
    '$' : 'Dólares',
    'Tonnes' : 'Toneladas'
})
print(df['Medida'].unique())
```

```python
df['Medida'] = df['Medida'].replace({
    'All': 'Todos(as)',
    'Value': 'Valor',
    'Weight': 'Peso'
})

print(df['Medida'].unique())
```

```python
df['Dia da Semana'] = df['Dia da Semana'].replace({
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo',
})

print(df['Dia da Semana'].unique())
```

```python
df['Direção'] = df['Direção'].replace({
    'Imports': 'Importações',
    'Exports': 'Exportações',
    'Reimports': 'Reimportações',

  })

print(df['Direção'].unique())
```

```python
colunas = ['Ano', 'País',
           'Mercadoria', 'Modo de Transporte',
           'Medida', 'Valor','Direção', 'Dia da Semana']
```

```python
for coluna in colunas:
    print(f"Valores únicos na coluna {coluna}")
    print(df[coluna].value_counts())
```

```python
import numpy as np
```

```python
linhas_nulas = np.any(pd.isnull(df) ,axis=1)

print(f'Número de linhas nulas: {np.sum(linhas_nulas)}')

df[linhas_nulas].head()
```

```python
# Porcentagem de valores nulos por coluna
nulos_por_coluna = df.isnull().sum() / len(df) * 100
print("Porcentagem de nulos por coluna:\n", nulos_por_coluna)

# Colunas com um único valor
colunas_constantes = [col for col in df.columns if len(np.unique(df[col].dropna())) == 1]
print("Colunas constantes (provavelmente inúteis):", colunas_constantes)
```

```python
df_limpo = df.dropna()
```

```python
df_limpo.isnull().sum()
```

```python
import plotly.express as px
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Calcular média e ordenar
valor_gasto = (
    df_limpo.groupby('País')['Valor']
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

# Criar gráfico
fig = px.bar(
    valor_gasto,
    x='País', color = 'País',
    y='Valor',
    title='Valor gasto no total em USD(Dólares)',
    labels={
        'País': 'País',
        'Valor': 'Valor Gasto no Total(Dólares)' # Adjusted label as Medida can be Dólares or Toneladas
    }
,
    color_discrete_map={
        'Brasil': 'green',
        'Estados Unidos': 'blue',
        'Alemanha': 'red',
        'Japão': 'purple',
        # adicione todos os países que quiser
    }
)

fig.show()
```

```python
# Calcular a contagem e ordenar
type_transporte = (
    df_limpo.groupby('País')['Modo de Transporte']
    .count()
    .sort_values(ascending=False)
    .reset_index()
)

# Criar gráfico (igual ao antigo, com cor por país)
fig = px.bar(
    type_transporte,
    x='País',
    y='Modo de Transporte',
    color='País',   # <<< ESSA LINHA faz igual ao gráfico anterior
    title='Modo de Transporte por País',
    labels={
        'País': 'País',
        'Modo de Transporte': 'Contagem de Modos de Transporte'
    }
)

fig.show()
```
